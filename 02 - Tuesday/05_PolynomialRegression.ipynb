```python
X = pd.DataFrame(np.c_[boston['LSTAT'], boston['RM']], columns=['LSTAT', 'RM'])
Y = boston['MEDV']
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print(X_train[0:5])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
y_train_predict = lin_model.predict(X_train)
r2_train = r2_score(Y_train, y_train_predict)
rmse_train = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
```

```python
y_test_predict = lin_model.predict(X_test)
r2_test = r2_score(Y_test, y_test_predict)
rmse_test = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
```

```python
print(f"R2: Train - {r2_train} | Test - {r2_test}")
print(f"rmse: Train - {rmse_train} | Test - {rmse_test}")
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=5)
X_train_poly = poly_features.fit_transform(X_train)

print(poly_features.get_feature_names_out(['LSTAT', 'RM']))
```

```python
X_train_poly[0:5]
```

```python
polynomial_model = LinearRegression()
polynomial_model.fit(X_train_poly, Y_train)

print(f"Model Coeffs: {polynomial_model.coef_}")
print(f"Model Constant: {polynomial_model.intercept_}")
```

```python
y_train_predict = polynomial_model.predict(X_train_poly)
r2_train = r2_score(Y_train, y_train_predict)
rmse_train = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))

y_test_predict = polynomial_model.predict(poly_features.fit_transform(X_test))
r2_test = r2_score(Y_test, y_test_predict)
rmse_test = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))

print(f"R2: Train - {r2_train} | Test - {r2_test}")
print(f"rmse: Train - {rmse_train} | Test - {rmse_test}")
```
