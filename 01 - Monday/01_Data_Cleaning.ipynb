```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randn(8,7))
df[df > 0.9] = np.nan
df

print(df.isnull())

for column_name in df:
    df[column_name].fillna(df[column_name].mean(), inplace=True)

df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv('Sberbank Housing Market.csv')
df.shape
df.dtypes

from pandas.plotting import scatter_matrix

df_fewer_columns = df.iloc[:, 15:20:1]
scatter_matrix(df_fewer_columns, figsize=(12, 12))

import missingno as msno

df_some_columns = df.iloc[:, 0:30:1]
msno.matrix(df_some_columns)
msno.bar(df_some_columns)
```

```python
limitPer = len(df_some_columns) * .80
df_some_columns.dropna(axis='columns', thresh=limitPer, inplace=True)
msno.bar(df_some_columns)
```
