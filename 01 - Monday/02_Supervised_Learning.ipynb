```python
import pandas as pd

credit_data = pd.read_csv("german credit data unclean.csv")

credit_data[0:12]
```

```python
credit_data.isnull()
```

```python
num_nulls = credit_data.isnull().sum().sum()
num_nulls
```

```python
for col in credit_data:
    null_count = credit_data[col].isnull().sum()
    print(f"{null_count} nulls in column {col}")
```

```python
import missingno as msno

msno.bar(credit_data)
```

```python
credit_data["age"].fillna(credit_data["age"].mean(), inplace=True)
print(credit_data["age"].mean())
credit_data.dropna(inplace=True)
msno.bar(credit_data)
```

```python
cols = ['check_account_status', 'credit_history', 'purpose', 'savings_account', 'employment_duration', 'gender_marriage', 'other_debtors', 'property', 'other_plans', 'housing', 'job', 'telephone', 'foreign_worker']

for col in cols:
    one_hot_check = pd.get_dummies(credit_data[col], prefix=col)
    credit_data.drop(col, axis=1, inplace=True)
    credit_data = one_hot_check.join(credit_data)

credit_data[0:12]
```

```python
from sklearn.linear_model import LogisticRegression

X = credit_data.loc[:, 'foreign_worker_A201':'dependents']
Y = credit_data['default']

logModel = LogisticRegression(solver='liblinear')
logModel.fit(X, Y)
print(logModel.score(X, Y))


```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

logModel = logModel.fit(X_train, y_train)
predictions = logModel.predict(X_test)
predictions
print(logModel.score(X_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, predictions))
```

```python
print(logModel.intercept_[0])
print(*logModel.coef_[0], sep='\n')

print(X_test[0:1].to_csv(index=False, sep='\n', header=False))
print(logModel.predict_proba(X_test[0:1]))
```

```python
import pickle
model_filename = 'germ_cred_model.pkl'
the_file = open(model_filename, 'wb')
pickle.dump(logModel, the_file)
the_file.close()
```
